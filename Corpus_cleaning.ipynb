## Sentence perplexity

```python
from transformers import AutoTokenizer, AutoModelForMaskedLM, BertForMaskedLM

BASE_MODEL = "neuralmind/bert-base-portuguese-cased"

tokenizer = AutoTokenizer.from_pretrained(BASE_MODEL)
model = BertForMaskedLM.from_pretrained(BASE_MODEL)
```

```python
model.to('cuda:0')
model.device
```

```python
from copy import deepcopy
import numpy as np
from tqdm import tqdm
ppls = []
for sentence in tqdm(n_all_sentences):
  tokenized = tokenizer.encode(sentence, return_tensors="pt")
  tokenized_labels = deepcopy(tokenized)
  tokenized_labels[0][0] = -100
  tokenized_labels[0][-1] = -100
  outputs = model(input_ids=tokenized.to("cuda:0"),labels=tokenized_labels.to("cuda:0"))  
  # print(np.exp(outputs[0].detach().numpy()))
  ppls.append(np.exp(outputs[0].cpu().detach().numpy()))
```

```python
df['ppls'] = ppls
df.head()
```

```python
df.to_csv("./clean_corpus.csv")
```

## Proceed

### Remove offensive

```python
not_offensive = df[df['offensive']==0]
not_offensive.info()
```

### Remove higher perplexities

```python
from scipy import stats
import math
import numpy as np

alpha = 0.25

mean = np.mean(not_offensive['ppls'])
std = np.std(not_offensive['ppls'])

z = stats.norm.ppf(1-alpha)

x = (z*std)+mean
x
```

```python
over_df = not_offensive[not_offensive['ppls'] >= x]
over_df.info()
```

```python
over_df
```

```python
not_offensive_lower_perplexities = not_offensive[not_offensive['ppls'] < x]
not_offensive_lower_perplexities.info()
```

### Remove big setences

```python
print("Mean:",not_offensive_lower_perplexities["lens"].mean())
print("Std:",not_offensive_lower_perplexities["lens"].std())
print("Min:",not_offensive_lower_perplexities["lens"].min())
print("Max:",not_offensive_lower_perplexities["lens"].max())
print("Mode:",not_offensive_lower_perplexities["lens"].mode())
```

```python
not_offensive_lower_perplexities_no_large = not_offensive_lower_perplexities[not_offensive_lower_perplexities['lens']<=11]
not_offensive_lower_perplexities_no_large.info()
```
